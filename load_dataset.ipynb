```python
import os
```

```python
base_path = "drive/MyDrive/Sketch2Face/dataset"
```

```python
sketch_dir = os.path.join(base_path, "sketches")
photo_dir  = os.path.join(base_path, "photos")

print("Sketch images:", len(os.listdir(sketch_dir)))
print("Photo images :", len(os.listdir(photo_dir)))
```

```python
def normalize_sketch_name(name):
    name = name.lower()

    # remove _sz1
    name = name.replace("-sz1", "")

    # replace M2 → m and F2 → f
    if name.startswith("m2"):
        name = name.replace("m2", "m", 1)
    if name.startswith("f2"):
        name = name.replace("f2", "f", 1)

    return name
```

```python
photo_files = os.listdir(photo_dir)
photo_dict = {p.lower(): p for p in photo_files}
```

```python
paired = []
missing = []
```

```python
for sketch in os.listdir(sketch_dir):
    normalized_name = normalize_sketch_name(sketch)

    if normalized_name in photo_dict:
        paired.append((sketch, photo_dict[normalized_name]))
    else:
        missing.append(sketch)
```

```python
print("Total paired samples:", len(paired))
print("Total missing sketches:", len(missing))
```
